2. Dealing with Outliers

Z-score method

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df1))

# Define a threshold for identifying outliers
threshold = 3

# Identify outliers
outliers_z = np.where(z_scores > threshold)
outliers_df_z = df1[(z_scores > threshold).any(axis=1)]

print("Outliers based on Z-score method:")
outliers_df_z
```

IQR method

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers_iqr = df1[((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Outliers based on IQR method:")
outliers_iqr
```

3. Coverting numeric values into categorical values

```python
import pandas as pd

# Example DataFrame (replace with your actual DataFrame loading code)
# Assuming df1 is already filled with NaN values replaced by means
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [0.5, 1.5, 2.5, pd.NA, 4.5],
    'C': [10, pd.NA, 30, 40, 50]
})

# Fill NaN values with the mean of each column
df1 = df.fillna(df.mean())

# Convert numeric values to categorical values
# Define bins for each numeric column
bins = {
    'A': [0, 2, 4, 6],
    'B': [0, 1, 3, 5],
    'C': [0, 20, 40, 60]
}

# Convert numeric columns to categorical using cut
for col in df1.columns:
    if pd.api.types.is_numeric_dtype(df1[col]):
        df1[col] = pd.cut(df1[col], bins=bins[col], labels=['Low', 'Medium', 'High'])

print("DataFrame with numeric values converted to categorical:")
print(df1)
```
