2. Dealing with Outliers

Z-score method

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df1))

# Define a threshold for identifying outliers
threshold = 3

# Identify outliers
outliers_z = np.where(z_scores > threshold)
outliers_df_z = df1[(z_scores > threshold).any(axis=1)]

print("Outliers based on Z-score method:")
outliers_df_z
```

IQR method

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers_iqr = df1[((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Outliers based on IQR method:")
outliers_iqr
```

Using SMOTE Oversampling

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

df1 = pd.read_csv('/content/creditcard.csv')

# Copy the data
df4 = df1.copy()

# Separate features (only numeric columns) and target
numeric_cols = df4.select_dtypes(include='number').columns
X = df4[numeric_cols]
y = df4['Class']  # Assuming 'Class' is the column indicating fraud (1) or not (0)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check the distribution after resampling
plt.figure(figsize=(8, 6))
sns.countplot(x=y_train_resampled)
plt.title('Class Distribution after SMOTE')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Print the value counts to verify
print("Value counts after SMOTE:")
print(y_train_resampled.value_counts())
```

Randomsampling for Undersampling

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Load your dataset (replace '/path/to/your/data.csv' with your actual file path)
df1 = pd.read_csv('/content/creditcard.csv')

# Copy the data
df5 = df1.copy()

# Separate features (only numeric columns) and target
numeric_cols = df5.select_dtypes(include='number').columns
X = df5[numeric_cols]
y = df5['Class']  # Assuming 'Class' is the column indicating fraud (1) or not (0)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Perform Random Undersampling
undersampler = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train, y_train)

# Check the distribution after undersampling
plt.figure(figsize=(8, 6))
sns.countplot(x=y_train_resampled)
plt.title('Class Distribution after Random Undersampling')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Print the value counts to verify
print("Value counts after Random Undersampling:")
print(y_train_resampled.value_counts())
```

6. Dimensionality Reduction

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

# Load your dataset
df1 = pd.read_csv('/content/creditcard.csv')

# Assuming 'Class' is the column indicating fraud (1) or not (0)
X = df1.drop('Class', axis=1)  # Features
y = df1['Class']  # Target

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Apply SMOTE to the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Standardize the data before applying PCA
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test)

# Apply PCA
pca = PCA(n_components=0.95, random_state=42)  # Retain 95% of variance
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Print the number of components selected by PCA
print(f"Number of components selected by PCA: {pca.n_components_}")

# Display the explained variance ratio
print(f"Explained variance ratio by PCA components:\n{pca.explained_variance_ratio_}")

# Visualize the cumulative explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance Ratio by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# Filter out numeric columns for X
numeric_cols = df10.select_dtypes(include='number').columns
X = df10[numeric_cols].drop('Class', axis=1)  # Exclude the target variable

# Define the target variable
y = df10['Class']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data (KNN is sensitive to the scale of the features)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)  # Start with a simple choice of k

# Train the model
knn.fit(X_train_scaled, y_train)

# Make predictions
y_pred = knn.predict(X_test_scaled)

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
